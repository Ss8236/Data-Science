### Project Summary: Exploratory Data Analysis of Amazon Prime Video Content Library

#### Overview
This project focuses on an Exploratory Data Analysis (EDA) of the Amazon Prime Video content library, utilizing two datasets: `titles.csv` (over 9,000 unique titles) and `credits.csv` (over 124,000 cast and crew entries). The datasets, specific to the United States, include metadata such as title, show type (movie or TV show), genres, release year, IMDb scores, TMDB popularity, runtime, age certification, production countries, and more. Conducted on June 13, 2025, this analysis leverages Python libraries—Pandas for data manipulation, NumPy for numerical operations, and Matplotlib/Seaborn for visualizations—to derive actionable insights for Amazon Prime Video’s content strategy in the competitive streaming industry.

####  Objective
The primary objective of this project is to analyze the Amazon Prime Video content library to uncover insights on content diversity, regional availability, temporal trends, and user engagement metrics (e.g., IMDb ratings, TMDB popularity). The goal is to empower Amazon Prime’s leadership to enhance commercial performance by optimizing content acquisition, expanding market reach, boosting user engagement, strengthening competitive positioning, and maximizing ROI on content investments. Specifically, the analysis aims to identify high-performing genres, prioritize production partnerships in key markets, promote high-rated content for user retention, ensure a fresh library to meet audience demands, and allocate budgets toward impactful content to drive profitability and subscription growth.

#### Methodology
The project follows a structured approach based on the provided project architecture: **Project Statement**, **Visualizations**, and **Drawing Insights**. The analysis adheres to the UBM rule (Univariate, Bivariate, Multivariate) to ensure comprehensive exploration.

1. **Project Statement**: The initial phase involved understanding the dataset and listing assumptions, such as IMDb scores reflecting user preferences and content diversity impacting engagement. Additional ideas were generated, including analyzing movies vs. TV shows, runtime correlations, and frequent actors/directors.

2. **Visualizations**: Twenty charts were created to address key questions:
   - **Univariate Analysis**: Histograms (e.g., IMDb scores, release years), bar charts (e.g., show types, genres), and pie charts (e.g., age certifications) revealed distributions. For instance, most titles are rated 6–8 on IMDb, movies outnumber TV shows, and drama/comedy genres dominate.
   - **Bivariate Analysis**: Boxplots (e.g., IMDb scores by show type), scatter plots (e.g., IMDb score vs. runtime), and heatmaps (e.g., genre co-occurrence) explored relationships. Movies generally have higher IMDb scores than TV shows, and drama often pairs with romance.
   - **Multivariate Analysis**: Scatter plots with hues (e.g., IMDb score vs. runtime by show type), facet grids, and violin plots analyzed complex interactions. US movies consistently score higher than TV shows, while Indian content shows more variance.

3. **Drawing Insights**: Insights were synthesized to answer key questions:
   - **Content Diversity**: Drama and comedy dominate, but high-rated genres like documentaries are underrepresented, offering growth opportunities.
   - **Regional Availability**: The US and India lead in volume, but smaller regions like Denmark have higher average ratings, suggesting quality-focused partnerships.
   - **Trends Over Time**: Most content is from 2000–2020, aligning with user preferences for fresh titles.
   - **User Engagement**: Movies generally have higher ratings, and high IMDb scores often align with TMDB popularity, though niche content exists.

#### Key Findings and Business Impact
The analysis revealed several actionable insights. Amazon Prime can enhance its content strategy by investing in high-rated genres like documentaries, focusing on optimal runtimes (90–120 minutes) for movies, and partnering with high-scoring regions like Denmark. Promoting high-rated movies and leveraging popular genre pairs (e.g., drama-romance) can boost engagement and retention. However, risks include potential churn from TV show fans due to lower ratings, missed opportunities in niche genres, and limited global appeal if regional diversity isn’t addressed.

## 3. ***Data Wrangling***

```python
# Filling the missing value in 'description','age_certification','seasons','imdb_score','tmdb_score'
titles_df['description'] = titles_df['description'].fillna('No description available')
titles_df['age_certification'] = titles_df['age_certification'].fillna('Unknown')
titles_df['seasons'] = titles_df['seasons'].fillna(0).astype('int64')
titles_df['imdb_score'] = titles_df['imdb_score'].fillna(titles_df['imdb_score'].median())
titles_df['tmdb_score'] = titles_df['tmdb_score'].fillna(titles_df['tmdb_score'].median())
#Checking the null data
titles_df.isnull().sum()
```

 **The missing values in the 'titles' dataset has been fixed**
  - missing values in 'description' has been filled with 'No Description availiable'.
  - missing values in 'age ceritification' has been fixed with 'Unknown'.
  -missing values in 'seasons' has been fixed with '0' and it's data type has been changed to integers.
  -missing values in 'imdb_score' has been fixed by putting median value of 'imdb_score'.
  -missing value in 'tmdb_score' has been fixed by putting median value of 'tmdb_score'.
  

```python
# Fixing Data types
import ast
titles_df['type'] = titles_df['type'].astype('category')
titles_df['age_certification'] = titles_df['age_certification'].astype('category')
titles_df['genres'] = titles_df['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
titles_df['production_countries'] = titles_df['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
credits_df['role'] = credits_df['role'].astype('category')
```

The initial data preprocessing steps involve transforming specific columns in the titles_df and credits_df DataFrames to ensure they are in the appropriate format for analysis.

*     First, the type and age_certification columns in titles_df are converted to categorical data types using the .astype('category') method, which optimizes memory usage and facilitates efficient grouping and filtering operations for these categorical variables.

*     Next, the genres and production_countries columns in titles_df, which are stored as string representations of lists, are parsed into actual Python lists using ast.literal_eval().

*     This transformation is applied conditionally with pd.notnull() to handle missing values, ensuring that NaN entries are replaced with empty lists ([]) to maintain consistency during subsequent analysis, such as exploding these columns for genre or country-based aggregations.

*     Finally, in the credits_df DataFrame, the role column is converted to a categorical data type to streamline operations involving the cast and crew roles (e.g., ACTOR or DIRECTOR).

These preprocessing steps lay the foundation for accurate and efficient exploratory data analysis by ensuring the data is structured appropriately for visualization and statistical computations.

```python
#Chart 11: Heatmap of Genre Co-Occurrence (Categorical-Categorical)
# Create a co-occurrence matrix for genres

# Explode the genres list to get individual genres and then find the top 10
genres_exploded = titles_df.explode('genres')
top_genres = genres_exploded['genres'].value_counts().head(10).index

# Filter the original titles_df to include only titles with more than one genre
titles_with_multiple_genres = titles_df[titles_df['genres'].apply(lambda x: len(x) > 1)]

# Create a co-occurrence matrix initialized with zeros
co_occurrence = pd.DataFrame(0, index=top_genres, columns=top_genres)

# Populate the co-occurrence matrix
for index, row in titles_with_multiple_genres.iterrows():
    genre_list = row['genres']
    # Iterate through each pair of genres within the list for the current title
    for g1 in genre_list:
        for g2 in genre_list:
            # Ensure both genres are in the top_genres list and are not the same
            if g1 in top_genres and g2 in top_genres and g1 != g2:
                co_occurrence.loc[g1, g2] += 1

plt.figure(figsize=(10, 8))
sns.heatmap(co_occurrence, annot=True, cmap='Blues', fmt='d')
plt.title('Genre Co-Occurrence Heatmap')
plt.show()
```

- **Why Chosen**: A heatmap visualizes relationships between categorical variables, here showing the frequency of genre co-occurrences for the top 10 genres on Amazon Prime.
- **Insights**: Drama frequently pairs with comedy (1,145 titles) and thriller (1,148 titles), while action and thriller (631 titles) and horror and thriller (554 titles) are common combinations, indicating popular genre pairings.
- **Positive Business Impact**: Producing content with high-frequency genre pairs like drama-comedy or action-thriller can attract broader audiences, enhancing viewership and engagement.
- **Negative Growth Risk**: Over-focusing on common genre pairs may stifle innovation, potentially overlooking unique combinations (e.g., family-documentary, 28 titles) that could appeal to niche audiences and diversify the content library.
